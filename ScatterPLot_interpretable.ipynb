```python
# importing Libraries
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#loading the data
df = pd.read_hdf('brfss.hdf5')
df.head(2)
```

```python
# extracting our x and y  features
x = df['HTM4']
y = df['WTKG3']
```

```python
# Normally plotting X and y
plt.plot(x, y, 'o');
plt.xlabel("plt.plot(x,y,'o')");
plt.title("Normally plotting");
```

```python
#add random noises to fill spaces between points
x_jitter = x + np.random.normal(0, 2, size=len(x))
y_jitter = y + np.random.normal(0, 2, size=len(y))
```

```python
#plotting with low marker size and zooming in
plt.plot(x_jitter, y_jitter, 'o',markersize=0.8,alpha = 0.08);
plt.xlabel("plt.plot(x_jitter, y_jitter, 'o',markersize=0.8,alpha = 0.08);");
plt.xlim([140,200]);
plt.ylim([20,150]);
plt.title('Plotting with a low markersize and transperant jittering points');
```

```python
# comparing the two plots
plt.rcParams["figure.figsize"] = (20,4)

plt.subplot(1,2,1)

plt.plot(x, y, 'o');
plt.xlabel("plt.plot(x,y,'o')", size = 15);
plt.title("Normally plotting", size = 15);

plt.subplot(1,2,2)

plt.plot(x_jitter, y_jitter, 'o',markersize=0.8,alpha = 0.08);
plt.xlabel("plt.plot(x_jitter, y_jitter, 'o',markersize=0.8,alpha = 0.08);",size = 15);
plt.xlim([140,200]);
plt.ylim([20,150]);
plt.title('Plotting with a low markersize and transparent jittering points',size = 15);
```
